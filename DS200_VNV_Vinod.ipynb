```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_table('All_India_Index_july2019_20Aug2020_dec20_1_0.csv', sep=',')
```

```python
df.describe()
```

## Scatterplot

We will plot two variables using scatter plot to understand correlation. Let's keep "Fuel And Light" in X axes and Plot two figure with Y ('Personal care and effects', 'Transport and communication'). As can be seen from figure below, variable 'Transport and communication' is tightly correlated with 'Fuel And Light'.

```python
sns.scatterplot(x = 'Fuel and light', y = 'Transport and communication', data=df, hue='Sector')
```

```python
sns.scatterplot(x = 'Fuel and light', y = 'Personal care and effects', data=df, hue='Sector')
```

## Bar Plot
Below figures shows number of data points in different price range for fruits.

```python
plt.figure(figsize=(15,4))
df['fruit_index_range'] = pd.cut(df['Fruits'], 5)
sns.barplot(x='fruit_index_range', y = 'Fruits', data=df)
```

## Box Plot.
We can see from figure that Vegetables index varies higest in November/December and lowest in July/August. July month has some outliers index values as well. Whereas August Does not have.

```python
plt.figure(figsize=(15,4))
sns.boxplot(x='Month', y='Vegetables', data=df)
```
